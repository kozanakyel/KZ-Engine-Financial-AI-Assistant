```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
range_list = list(range(10,21,10))
ind_list = ['rsi', 'mfi']
```

```python
def plot_indicators(r_list: list, indicator: str, prompt_no: int):
    col_list = [f'{indicator}_{x}' for x in r_list]
    prompt_df = pd.read_csv(f'model_fine_tuned_data/prompt_{prompt_no}.csv', index_col=[0])
    prompt_df.plot(y=col_list, kind='line', figsize=(10,3))
    plt.xticks(rotation=10, fontsize=8)
    return prompt_df
```

```python
dt = plot_indicators(range_list, 'rsi', 5)
```

```python
dt = plot_indicators(range_list, 'mfi', 5)
```

```python
# datatset.json file for each line of prompt format
line_dataset = {"prompt":f"{dt.to_json()}:", "completion": " RSI_5 indikatoru ani bir ivme ile 80 seviyesi civarlarina yukselmistir.    bu ani hareketler onem arz etmektedir. MFI_5 indikatoru 60 seviyelerine duserken MFI_10 indikatoru 70 seviyelerine yaklasmkatadir. Bu da orta vadede iyi bir yukselis gelebilecegi anlamina gelmektedir. EMA_5 indikatoru de tekrar yukselisine devam etmektedir."}

with open('dataset.json', 'a+') as j:
    j.write('\n')
    j.write(json.dumps(line_dataset))
```
